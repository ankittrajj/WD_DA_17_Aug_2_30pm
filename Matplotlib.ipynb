```python
# lineplot
# histogram
# Bar chart
# pie chart
# subplot
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
days = np.array([1,2,3,4,5])
temp = np.array([34,23,12,43,32])
plt.xlabel('Days')
plt.ylabel("Temp.")
plt.title('Days vs Temp Graph')
plt.plot(days,temp,color = 'r')
```

```python
days = np.array([1,2,3,4,5])
temp1 = np.array([34,23,12,43,32])
temp2 = [34,12,56,3,45]
plt.xlabel('Days')
plt.ylabel("Temp.")
plt.title('Days vs Temp Graph')
plt.plot(days,temp1,color = 'r')
plt.plot(days,temp2)
```
